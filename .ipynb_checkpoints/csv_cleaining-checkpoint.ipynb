# CSV cleaning

## Imports

```python
import pandas as pd
import numpy as np

import os

import re

import nltk

import spacy
from spacy_lefff import LefffLemmatizer, POSTagger
```

## Read data

```python
path = 'data'
df1 = pd.read_csv(os.path.join(path, 'tweets.csv'))
df2 = pd.read_csv(os.path.join(path, 'tweets_2.csv'))
df3 = pd.read_csv(os.path.join(path, 'tweets_3.csv'))

df11 = pd.read_csv(os.path.join(path, 'retweets.csv'))
df22 = pd.read_csv(os.path.join(path, 'retweets_2.csv'))
df33 = pd.read_csv(os.path.join(path, 'retweets_3.csv'))
```

```python
## remove duplicates due to how data was extracted
to_rmv = df1[df1['author_id'] == 398785825]['status_id']
df11 = df11[~df11['tweet_id'].isin(to_rmv)]
```

```python
df = pd.concat([df1, df2, df3], ignore_index = True)
dff = pd.concat([df11, df22, df33], ignore_index = True)
```

```python
dfh = pd.read_csv(os.path.join(path, 'tweets_hold_up_1.csv'))
dfhh = pd.read_csv(os.path.join(path, 'retweets_hold_up_1.csv'))
```

```python
## removing all the tweets that were already there
duplicates_status_ids = dfh['status_id'][dfh['status_id'].isin(df['status_id'])]
dfh = dfh[~dfh['status_id'].isin(duplicates_status_ids)]
dfhh = dfhh[dfhh['tweet_id'].isin(duplicates_status_ids)]
```

```python
tweets = pd.concat([df, dfh], ignore_index = True)
retweets = pd.concat([dff, dfhh], ignore_index = True)
```

```python
len(retweets)
```

```python
def rmv_dots(text):
    if text[len(text)-1] == '…':
        text = text[:-1]
    if text[:2] == 'RT':
        text = text.split(':', 1)[1][1:]
        
    return text

def rmv_char(text):
    ## removes all unwanted characters of the text
    text = re.sub("""[\n´]""", " ", text)
    text = re.sub("""["'+"*%&/()"=¦@#°§¬|¢\[\]\-\_\—‘’“”•`\^{}~¥©?.,:!$;«»<>]""", "", text)
    return re.sub("""\d""", " ", text)

def rmv_hyperlink(text):
    ## removes all hyperlinks of the text
    return re.sub(r'''(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'".,<>?«»“”‘’]))''', " ", text)

def lemmatize(serie):
    pos = POSTagger()
    french_lemmatizer = LefffLemmatizer(after_melt = True)
    
    nlp = spacy.load('fr_core_news_sm')
    nlp.add_pipe(pos, name = 'pos', after = 'parser')
    nlp.add_pipe(french_lemmatizer, name = 'lefff', after = 'pos')
    
    lemmatized = serie.map(
        lambda x : [doc.lemma_ for doc in nlp(x)]
    )
    return lemmatized

def clean_tweets(serie):
    serie = serie.map(rmv_hyperlink)
    serie = lemmatize(serie)
    serie = serie.map(lambda x : " ".join(x))
    serie = serie.map(lambda x : x.lower()).map(rmv_char)
    serie = serie.map(nltk.word_tokenize)
    stopwords = nltk.corpus.stopwords.words('french')
    serie = serie.map(lambda tokens : [x for x in tokens if x not in stopwords])
    serie = serie.map(lambda tokens : [x for x in tokens if x.isalpha()])
    return serie.map(lambda x : " ".join(x))
```

```python
bodies = tweets['body']
bodies = bodies.map(
    rmv_dots
)
```

```python
bodies = clean_tweets(bodies)
```

```python
tweets['body'] = bodies
```

```python
tweets.to_csv(os.path.join('data', 'clean_tweets.csv'))
retweets.to_csv(os.path.join('data', 'clean_retweets.csv'))
```
